```python
from scipy.stats import f_oneway

class_A = [85, 90, 88, 82, 87]
class_B = [76, 78, 80, 81, 75]
class_C = [92, 88, 94, 89, 90]


f_stat, p_value = f_oneway(class_A, class_B, class_C)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)


if p_value < 0.05:
    print("Reject Null Hypothesis → Significant difference in means")
else:
    print("Fail to Reject Null → No significant difference in means")
```
