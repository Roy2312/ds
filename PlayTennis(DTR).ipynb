```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


df = pd.read_csv("play_tennis.csv")


df.drop("Date", axis=1, inplace=True)


le_outlook = LabelEncoder()
le_wind = LabelEncoder()
le_play = LabelEncoder()

df['Outlook'] = le_outlook.fit_transform(df['Outlook'])
df['Wind'] = le_wind.fit_transform(df['Wind'])
df['PlayTennis'] = le_play.fit_transform(df['PlayTennis'])


X = df[['Outlook', 'Wind']]
y = df['PlayTennis']


model = DecisionTreeClassifier()
model.fit(X, y)


plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['Outlook', 'Wind'], class_names=['No', 'Yes'], filled=True)
plt.show()

sample = [[le_outlook.transform(['Sunny'])[0], le_wind.transform(['Weak'])[0]]]
prediction = model.predict(sample)

print("Prediction:", le_play.inverse_transform(prediction))

```
