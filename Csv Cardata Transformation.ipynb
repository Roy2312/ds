```python
import pandas as pd


df = pd.read_csv("Fs Car_data.csv")


high_price_cars = df[df['Sell Price'] > 4000]
print("Cars with Sell Price > 4000:")
print(high_price_cars)

sorted_df = df.sort_values(by='Sell Price')
print("\nSorted Cars by Sell Price:")
print(sorted_df)


grouped = df.groupby('Make')
print("\nGrouped by Make:")
print(grouped.size())

avg_sell_price = grouped['Sell Price'].mean()
print("\nAverage Sell Price by Make:")
print(avg_sell_price)
```
