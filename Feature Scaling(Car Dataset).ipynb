```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = pd.read_csv("Fs Car_data.csv")


df["Mileage"] = df["Mileage"].astype(str).str.replace(",", "")
df["Mileage"] = df["Mileage"].astype(int)


numerical_features = df[["Mileage", "Sell Price"]]


standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(numerical_features)
standard_scaled_df = pd.DataFrame(standard_scaled, columns=["Mileage_Standardized", "SellPrice_Standardized"])


minmax_scaler = MinMaxScaler()
normalized = minmax_scaler.fit_transform(numerical_features)
normalized_df = pd.DataFrame(normalized, columns=["Mileage_Normalized", "SellPrice_Normalized"])


result_df = pd.concat([df, standard_scaled_df, normalized_df], axis=1)
print(result_df)
```
